```python
# Create a Map instance
#     map_center = [44.9778, -93.2650]
m = folium.Map(location=[44.9778, -93.2650],
    zoom_start=12, control_scale=True)

# Add marker
# Run: help(folium.Icon) for more info about icons
folium.Marker(
    location=[44.95897, -93.27782],
    popup='McD',
    icon=folium.Icon(color='red', icon="info-sign"),
).add_to(m)

folium.Marker(
    location=[44.99878, -93.2854],
    popup='Taco Bell',
    icon=folium.Icon(color='purple', icon='bell'),
).add_to(m)

folium.Marker(
    location=[44.97393, -93.22494],
    popup='Burger King',
    icon=folium.Icon(color='orange', icon='info-sign'),
).add_to(m)

folium.Marker(
    location=[44.97543, -93.27484],
    popup='Subway',
    icon=folium.Icon(color='green', icon='info-sign'),
).add_to(m)

folium.Marker(
    location=[45.05787, -93.24828],
    popup='Sonic',
    icon=folium.Icon(color='pink', icon='info-sign'),
).add_to(m)

folium.Marker(
    location=[44.97387, -93.22957],
    popup='Chick Fil A',
    icon=folium.Icon(color='darkred', icon='thumbs-up'),
).add_to(m)

Geocoder().add_to(m)

#Show map
m
```
