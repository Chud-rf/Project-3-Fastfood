```python
# Load CSVs
menu_df = pd.read_csv('Data/fastfood.csv')

chains_df = pd.read_csv('Data/chains.csv')
```

```python
menu_df.info()
```

```python
# Drop Columns missing several values
# Dropping Salad because it's not a useful field
menu_df = menu_df.drop(labels=['vit_a', 'vit_c', 'calcium', 'salad'], axis=1)
menu_df.info()
```

```python
# Clean Fiber and Protien missing values
# Calculate column averages
fiber_avg = menu_df['fiber'].mean()
protein_avg = menu_df['protein'].mean()

# Fill missing values with averages
menu_df['fiber'].fillna(fiber_avg, inplace=True)
menu_df['protein'].fillna(protein_avg, inplace=True)

# Verify the changes
menu_df.info()
```

```python
numeric_columns = menu_df[['calories', 'cal_fat', 'total_fat', 'sat_fat', 'trans_fat', 'cholesterol', 'sodium', 'total_carb', 'fiber', 'sugar', 'protein']]

# chain_nutrition = menu_df.groupby('restaurant').mean()
chain_nutrition = numeric_columns.groupby(menu_df['restaurant']).mean()

chain_nutrition.reset_index(inplace=True)

chain_nutrition
```

```python
import matplotlib.pyplot as plt

def draw_calorie_circles(item1_calories, item2_calories):
    # Set up figure with two subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))

    # Draw circle for Item 1
    circle1 = plt.Circle((0.5, 0.5), 0.4, edgecolor='white', facecolor='lightblue')
    axs[0].add_patch(circle1)
    axs[0].text(0.5, 0.5, str(item1_calories), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[0].set_title('Item 1 - Calorie Count')
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    axs[0].spines['top'].set_visible(False)
    axs[0].spines['right'].set_visible(False)
    axs[0].spines['bottom'].set_visible(False)
    axs[0].spines['left'].set_visible(False)

    # Draw circle for Item 2
    circle2 = plt.Circle((0.5, 0.5), 0.4, edgecolor='white', facecolor='lightcoral')
    axs[1].add_patch(circle2)
    axs[1].text(0.5, 0.5, str(item2_calories), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[1].set_title('Item 2 - Calorie Count')
    axs[1].set_aspect('equal', adjustable='box')
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    axs[1].spines['top'].set_visible(False)
    axs[1].spines['right'].set_visible(False)
    axs[1].spines['bottom'].set_visible(False)
    axs[1].spines['left'].set_visible(False)

    # Adjust layout to prevent overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage
item1_calories = 300
item2_calories = 200

draw_calorie_circles(item1_calories, item2_calories)
```

```python
import matplotlib.pyplot as plt

def draw_calorie_comparison(menu_df, item1, item2):
    # Mapping of fast-food chains to signature colors
    chain_colors = {
        'Mcdonalds': 'goldenrod',
        'Chick Fil-A': 'deepskyblue',
        'Sonic': 'red',
        'Arbys': 'maroon',
        'Burger King': 'orange',
        'Dairy Queen': 'purple',
        'Subway': 'forestgreen',
        'Taco Bell': 'yellow'
    }

    # Get data for the selected items
    item1_data = menu_df[menu_df['item'] == item1]
    item2_data = menu_df[menu_df['item'] == item2]

    if item1_data.empty or item2_data.empty:
        print(f"Error: Items not found in the menu data for comparison - {item1} or {item2}")
        return

    item1_data = item1_data.iloc[0]
    item2_data = item2_data.iloc[0]

    # Set up figure with two subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))

    # Draw circle for Item 1
    chain_name_1 = item1_data['restaurant']
    chain_color_1 = chain_colors.get(chain_name_1, 'gray')  # Default to gray if color not specified
    circle1 = plt.Circle((0.5, 0.5), 0.4, edgecolor='white', facecolor=chain_color_1)
    axs[0].add_patch(circle1)
    axs[0].text(0.5, 0.5, str(item1_data['calories']), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[0].set_title(f'{chain_name_1} - {item1} Calories')
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    axs[0].spines['top'].set_visible(False)
    axs[0].spines['right'].set_visible(False)
    axs[0].spines['bottom'].set_visible(False)
    axs[0].spines['left'].set_visible(False)

    # Draw circle for Item 2
    chain_name_2 = item2_data['restaurant']
    chain_color_2 = chain_colors.get(chain_name_2, 'gray')  # Default to gray if color not specified
    circle2 = plt.Circle((0.5, 0.5), 0.4, edgecolor='white', facecolor=chain_color_2)
    axs[1].add_patch(circle2)
    axs[1].text(0.5, 0.5, str(item2_data['calories']), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[1].set_title(f'{chain_name_2} - {item2} Calories')
    axs[1].set_aspect('equal', adjustable='box')
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    axs[1].spines['top'].set_visible(False)
    axs[1].spines['right'].set_visible(False)
    axs[1].spines['bottom'].set_visible(False)
    axs[1].spines['left'].set_visible(False)

    # Adjust layout to prevent overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage
# Replace 'Item1' and 'Item2' with the actual item names you want to compare
item1_name = 'Item1'
item2_name = 'Item2'
draw_calorie_comparison(menu_df, item1_name, item2_name)
```

```python
# Example usage with specific item names
item1_name = 'Big Mac'
item2_name = 'Chicken Sandwich'
draw_calorie_comparison(menu_df, item1_name, item2_name)
```

```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Define the chain-to-logo mapping
chain_logos = {
    'Mcdonalds': 'img/mcdonalds_logo.png',
    'Chick Fil-A': 'img/chickfila_logo.png',
    'Sonic': 'img/sonic_logo.png',
    'Arbys': 'img/arbys_logo.png',
    'Burger King': 'img/burgerking_logo.png',
    'Dairy Queen': 'img/dairyqueen_logo.png',
    'Subway': 'img/subway_logo.png',
    'Taco Bell': 'img/tacobell_logo.png',
}

def draw_calorie_comparison(menu_df, item1, item2):
    # Set up figure with two subplots
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))

    # Get data for the selected items
    item1_data = menu_df[menu_df['item'] == item1].iloc[0]
    item2_data = menu_df[menu_df['item'] == item2].iloc[0]

    # Draw circle for Item 1
    circle1 = plt.Circle((0.5, 0.5), 0.4, color='lightblue', edgecolor='white')
    axs[0].add_patch(circle1)
    axs[0].text(0.5, 0.5, str(item1_data['calories']), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[0].set_title(f'{item1_data["restaurant"]} - {item1_data["item"]}')
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].axis('off')

    # Draw circle for Item 2
    circle2 = plt.Circle((0.5, 0.5), 0.4, color='lightcoral', edgecolor='white')
    axs[1].add_patch(circle2)
    axs[1].text(0.5, 0.5, str(item2_data['calories']), color='black', ha='center', va='center', fontsize=12, fontweight='bold')
    axs[1].set_title(f'{item2_data["restaurant"]} - {item2_data["item"]}')
    axs[1].set_aspect('equal', adjustable='box')
    axs[1].axis('off')

    # Load and display chain logos
    for ax, chain_name in zip(axs[2:], [item1_data['restaurant'], item2_data['restaurant']]):
        logo_path = chain_logos.get(chain_name, 'img/default_logo.png')
        logo_img = plt.imread(logo_path)
        imagebox = OffsetImage(logo_img, zoom=0.2)
        ab = AnnotationBbox(imagebox, (0.5, 0.5), frameon=False, boxcoords="axes fraction", pad=0)
        ax.add_artist(ab)
        ax.axis('off')

    # Adjust layout to prevent overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage with specific item names
item1_name = 'Big Mac'
item2_name = 'Chicken Sandwich'
draw_calorie_comparison(menu_df, item1_name, item2_name)
```
